#**Problema #3:**

###**Juego de Cara o Sello**
Crea un programa que permita jugar a "Cara o Sello" con el usuario. El programa seleccionará de
manera aleatoria "cara" o "sello" y luego pedirá al usuario que haga su adivinanza. Si acierta, se le

mostrará al usuario la cantidad de puntos obtenidos y se le preguntará si quiere jugar de nuevo. Si el
usuario elige "sí", el juego se reinicia. Si elige "no", el programa finaliza.


#**ALGORITMO PARA JUEGO DE CARA O SELLO**

1. Iniciar

2.Se importa la biblioteca random, que será útil para seleccionar elementos aleatorios (como "cara" o "sello") más adelante.

3. **Definir variables**  
   - Crear una variable entera "puntos" y establecerla en 0 para contar los puntos del usuario.
   - Crear una variable entera "vidas" y establecerla en 3, que representará la cantidad de vidas disponibles para el usuario.

4. **Bucle principal**  
   Se inicia un bucle que se ejecutará mientras "vidas" sea mayor que 0.

5. **Solicitar adivinanza al usuario**  
   - Mostrar un mensaje al usuario pidiéndole que adivine: "Adivina: ¿cara o sello?" y almacenar su respuesta en la variable eleccion_usuario.

6.

   - Mostrar el resultado: "La moneda cayó en: "resultado".

7. **Comparar adivinanza con resultado**  
   - Si "eleccion_usuario" es igual a "resultado":

   - Sino:

8. - Si "vidas" es igual a 0, mostrar un mensaje: "Se acabaron tus vidas. Juego terminado." y romper el bucle.

```python
import random

opciones = ["cara", "sello"]
puntos = 0
vidas = 3

while vidas > 0:
    eleccion_usuario = input("Adivina: ¿cara o sello?: ")

    resultado = random.choice(opciones)
    print("La moneda cayó en: ", resultado)

    if eleccion_usuario == resultado:
        puntos += 1
        print("¡Acertaste! Puntos acumulados:", puntos)
    else:
        vidas -= 1
        print("Fallaste. Vidas restantes:", vidas)

    if vidas == 0:
        print("Se acabaron tus vidas. Juego terminado.")
        break
```
