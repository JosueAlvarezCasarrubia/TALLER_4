    Declare int puntos = 0

```python
import random

opciones = ["cara", "sello"]
puntos = 0
vidas = 3

while vidas > 0:
    eleccion_usuario = input("Adivina: ¬øcara o sello?: ")

    resultado = random.choice(opciones)
    print("La moneda cay√≥ en: ", resultado)

    if eleccion_usuario == resultado:
        puntos += 1
        print("¬°Acertaste! Puntos acumulados:", puntos)
    else:
        vidas -= 1
        print("Fallaste. Vidas restantes:", vidas)

    if vidas == 0:
        print("Se acabaron tus vidas. Juego terminado.")
        break
```
